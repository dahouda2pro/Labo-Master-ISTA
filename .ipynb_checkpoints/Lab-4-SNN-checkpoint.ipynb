# 5. Division de donnees en X et y

```python
X = data.iloc[ : , :-1]
y = data.iloc[ : , -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=56
)
```

```python
print("X-train:", X_train.shape)
print("X-test:", X_test.shape)
print("y-train:", y_train.shape)
print("y-test:", y_test.shape)
```

```python
X_train
```

```python
# Standardiser les valeurs
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

# 6. Construction de Reseau de Neuron

```python
# Creation du modele de Reseau de Neuron
model_snn = Sequential()
```

```python
# Ajout de la premiere couche
model_snn.add(Dense(16, input_shape=(22,), activation='relu'))

# Ajout de la couche de sortie
model_snn.add(Dense(1, activation='sigmoid'))

# Compiler le modele
model_snn.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='binary_crossentropy', 
    metrics=['accuracy'])

# Affichage de l'architecture du reseau de neuron
model_snn.summary()
```

```python
# Entrainment du modele
history = model_snn.fit(X_train, y_train, epochs=50, verbose=1, validation_data=(X_test, y_test))
```

```python
# Affichage de l'entrainement
epoch_accuracy = history.history['accuracy']
epoch_loss = history.history['loss']

epoch_val_accuracy = history.history['val_accuracy']
epoch_val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, label='Train Accuracy')
plt.plot(range(0, len(epoch_val_accuracy)), epoch_val_accuracy, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Entrainement du Reseau de Neuron')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(0, len(epoch_loss)), epoch_loss, label='Train loss')
plt.plot(range(0, len(epoch_val_loss)), epoch_val_loss, label='Test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Entrainement du Reseau de Neuron')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, label='Train Accuracy')
plt.plot(range(0, len(epoch_loss)), epoch_loss, label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy vs Loss')
plt.title('Entrainement du Reseau de Neuron')
plt.legend()
plt.show()
```

```python
# Evaluer
model_snn.evaluate(X_test, y_test)
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_predit_lr = model_lr.predict(X_test)
print("Accuracy de LR:", accuracy_score(y_test, y_predit_lr) * 100, "%")
print("MSE de LR:", mean_squared_error(y_test, y_predit_lr))
confusion_matrix(y_test, y_predit_lr)
```
